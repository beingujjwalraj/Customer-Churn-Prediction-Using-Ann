```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
#import the dataset
data=pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
#preprocessing the data
#drop the irrelevant colmns
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
#encode the categorical variable
label_encoder_gender=LabelEncoder()
data['Gender']=label_encoder_gender.fit_transform(data['Gender'])
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Assuming `data` is a pandas DataFrame containing the 'Geography' column
onehot_encoder_geo = OneHotEncoder()

# Fit and transform the 'Geography' feature
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])

# Convert the sparse matrix to a dense array
geo_encoded_array = geo_encoder.toarray()

# Get feature names
geo_feature_names = onehot_encoder_geo.get_feature_names_out(['Geography'])

# Create a DataFrame from the dense array with appropriate column names
geo_encoded_df = pd.DataFrame(geo_encoded_array, columns=geo_feature_names)

# Display the encoded DataFrame
geo_encoded_df
```

```python
#combine one hot encoder columns with the original data
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
data
```

```python
#saving the encoder and scaler as picke file
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)

```

```python
#divide the dataset into independent and dependent features
x=data.drop('Exited',axis=1)
y=data['Exited']
#split the data in training and testing set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#scale these features
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
scaled_data=x_test
```

```python
# Convert to DataFrame for better visualization
scaled_df = pd.DataFrame(scaled_data, columns=x.columns)
scaled_df
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python

# Building the ANN model
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train.shape[1],)),  # Hidden layer 1
    Dense(32, activation='relu'),                                  # Hidden layer 2
    Dense(1, activation='sigmoid')                                 # Output layer
])
```

```python
model.summary()
```

```python
import tensorflow as tf

# Define optimizer with learning rate
opt = tf.keras.optimizers.Adam(learning_rate=0.01)

# Compile the model
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#setup the tensorboard
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
log_dir='logs/fit/'+datetime.datetime.now().strftime('%d/%m/%y - %H:%M:%S')
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#setup earlystoppings
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
#train the model
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5')
```

```python
# Evaluate the model on test data
results = model.evaluate(x_test, y_test, verbose=1)

# Extract and print the accuracy
accuracy = results[1]  # The second element corresponds to 'accuracy' metric
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Make predictions
y_pred = (model.predict(x_test) > 0.5).astype("int32")  # Convert probabilities to binary predictions

# Generate a classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
